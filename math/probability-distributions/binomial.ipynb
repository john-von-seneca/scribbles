* Distribution of N independent Bernoulli veriables
* $\P{X=m \mid N, \mu} = \nck{N}{m} \mu^{m} (1-\mu)^{N-m}$
* Statistics:
\begin{array}{llr}
\E{X} &= \sum_{n=1}^{N} \E{\bern{X \mid \mu}} = N \mu\\
\E{X^2} &= \sum_{n=1}^{N} \E{\bern{X^2 \mid \mu}} = N \mu\\
\V{X} &= sum_{n=1}^{N} \V{\bern{X \mid \mu}} = N \mu (1-\mu)
\end{array}

```python
from operator import mul    # or mul=lambda x,y:x*y
from fractions import Fraction
from functools import reduce

def nCk(n,k): 
  return int( reduce(mul, (Fraction(n-i, i+1) for i in range(k)), 1) )

def show_binomial(n,p=0.25):
    x = range(n+1)
    y = [nCk(n, k)*(p**k)*((1-p)**(n-k)) for k in x]
    plt.plot(x, y)
    plt.title('Binomial Distribution')
    plt.show()

interact(show_binomial, n=(0,20),p=(0,1,0.1))
```
