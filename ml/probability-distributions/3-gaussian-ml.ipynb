Maximum Likelihood for Gaussian
==================

<div id='StandardForm' \>
Standard Form  
$$
\Nstdx
=
\multivarcoeff \exp \multivarexpstd
$$

Given a data set $\Xb = \left(\xb_1, \cdots, \xb_N \right)^T$ and each
$\xb_n$ be drawn independently from a Multivariate Gaussian given by
<a href='#StandardForm'>Standard Form </a>.
The Log Likelihood is then given by
$$
\ln p(\Xb \mid \mub, \Sigma)
=
-\frac{ND}{2} \ln(2\pi)
-\frac{N}{2} \ln \left|\Sigma\right|
-\frac{1}{2} \sum_{n=1}^{N} (\xb_n - \mub)^T \Sigma^{-1} (\xb_n - \mub)
$$

The derivative of the log likelihood is given by
$$
\frac{\partial}{\partial \mub} \ln p(\Xb \mid \mub, \Sigma)
=
\sum_{n=1}^{N} \Sigma^{-1} (\xb_n - \mub)
$$
Hence, the ML estimate of the mean is
$$
\mub_{ML} = \frac{1}{N} \sum_{n=1}^{N} \xb_n
$$

\citeme{Magnus and Neudecker 1999}  
Finding covariance is quite involved.  
The covariance is given by
$$
\Sigma_{ML} = \frac{1}{N} \sum_{n=1}^{N} (\xb_n - \mub)(\xb_n - \mub)^T
$$

Also, the expected values of these estimates are
$$
\arrthree{
\E{\mub_{ML}}
&=
\mub
\\
\E{\Sigma_{ML}}
&=
\frac{N-1}{N} \Sigma
}
$$

This is a biased estimate. So we set
$$
\tilde{\Sigma} = \fracone{N-1}
\sumnN (\xb_n - \ml{\mub}) (\xb_n - \ml{\mub})^T
$$
