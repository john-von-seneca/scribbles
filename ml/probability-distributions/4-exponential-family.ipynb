# Exponential family

<div id='ExpFamilyGeneralForm'/>
General form:
$$
p(\xb \mid \eta)
=
h(\xb) g(\eta)
\exp \left\{ \eta^T \ub(\xb) \right\}
$$
Here
* $\eta$ are the natural parameters of the distribution.
* $g(\eta)$ is the normalizing factor

Ex: Bernoulli
---------------
$$
\arrthree{
p(x \mid \mu)
&=
\mu^x (1-\mu)^{1-x}
\\
&=
\exp
\left\{
  x     \ln \mu +
  (1-x) \ln (1-\mu)
\right\}
\\
&=
(1-\mu)
\exp
\left\{
  \ln \left(\dfrac{\mu}{1-\mu}\right)x
\right\}
}
$$
Compare this with the general form of exponential familty, we have
$$\nu = \ln \left( \dfrac{\mu}{1-\mu} \right)$$
We obtain $\mu = \sigma(n)$ which is called logistic sigmoid function,
given by
$$
\sigma(\eta)
=
\dfrac{1}{1+\exp(-\eta)}
\\
\text{where }
\sigma(-\eta) = 1 - \sigma(\eta)
$$
Thus we can write Bernoulli as 
$$
\arrthree{
p(x \mid \mu)
&=
(1 - \sigma(\eta)) \exp(\eta x)
\\
&=
\sigma(-\eta) \exp(\eta x)
\\
\text{Thus} \\
u(x) &= x \\
h(x) &= 1 \\
g(\eta) &= \sigma(-\eta)
}
$$

Ex: Multinomial
--------------

Multinomial, for a single observation, $\xb$ takes the form
$$
\arrthree{
p(\xb \mid \mub)
&= \prod_{k=1}^{M} \mu_{k}^{x_k}
\\
&= \exp
\left\{
  \sum_{k=1}^{M} x_k \ln \mu_k
\right\}
\\
&= \exp(\etabt \xb)
& \commentgray{$\eta_k = \ln \mu_k$}
}
$$
Comparing with <a href='#ExpFamilyGeneralForm'>General Form of the exponential family</a>, we get
$$
\arrthree{
\ub(\xb) &= \xb \\
h(\xb) &= 1 \\
g(\etab) &= 1\\
}
$$

Since $\sum_{k=1}^{M} \mu_k = 1$, the #parameters is M-1. Thus
$$
\arrthree{
p(\xb \mid \mub)
&=
\EXP{\sumkM x_k \ln \mu_k}
\\
&=
\EXP{
  \sumkMl x_k \ln \mu_k
  + \left( 1 - \sumkMl x_k \right)
    \ln \left( 1 - \sumkMl \mu_k \right)
}
\\
&=
\EXP{
  \sumkMl x_k \ln \left( \dfrac{\mu_k}{1-\sumjMl \mu_j} \right)
  + \ln \left( 1 - \sumkMl \mu_k \right)
}
\\ &=
\left( 1 - \sumkMl \mu_k \right)
~\EXP{\sumkMl x_k \ln \left( \dfrac{\mu_k}{1-\sumjMl \mu_j} \right)}
}
$$

We now need to find the expresion of each parameter $\mu_k$

$$
\arrthree{
\eta_k &= \ln \left( \dfrac{\mu_k}{1-\sum_j \mu_j} \right)
\\
e^{\eta_k} &= \dfrac{\mu_k}{1-\sum_j \mu_j}
}
$$

we know that 
$$
\arrthree{
a &= \frac{b}{1-b} \\
\fracone{a} &= \frac{1-b}{b} = \fracone{b} - 1 \\
\frac{1+a}{a} &= \fracone{b} \\
\frac{a}{a+1} &= b
}
$$

Sum from k=1 to M-1 both sides.  

Note: all the summations are from 1 to M-1.

$$
\arrthree{
\dfrac{\sum_k \mu_k}{1-\sum_j \mu_j} &= \sum_k e^{\eta_k}
& \commentgray{Add 1 and flip}
\\
\sum_k \mu_k &= \dfrac{\sum_j e^{\eta_k}}{1 + \sum_j e^{\eta_k}}
}
$$

Hence
$$
\arrthree{
e^{\eta_k} &= \dfrac{\mu_k}{1-\sum_j \mu_j}
\\ &=
\dfrac{\mu_k}{1-\dfrac{\sum_j e^{\eta_k}}{1 + \sum_j e^{\eta_j}}}
\\ &=
\dfrac{\mu_k}{\dfrac{1}{1 + \sum_j e^{\eta_j}}}
\\
\dfrac{e^{\eta_k}}{1 + \sum_j e^{\eta_j}} &= \mu_k
\\
\mu_k &= \dfrac{e^{\eta_k}}{1 + \sum_j e^{\eta_j}}
}
$$
This is called the softmax function or the normalized exponential.

Thus
$$
\arrthree{
\left( 1 - \sumkMl \mu_k \right) 
&=
1 - \displaystyle \sum_{k}^{M-1} \dfrac{e^{\eta_k}}{1 + \displaystyle \sum_j^{M-1} e^{\eta_j}}
\\ &=
1 - \dfrac{\displaystyle \sum_{k}^{M-1} e^{\eta_k}}{1 + \displaystyle \sum_j^{M-1} e^{\eta_j}}
\\ &=
\fracone{1 + \displaystyle \sum_j^{M-1} e^{\eta_j}}
}
$$

$$
\arrthree{
p(\xb \mid \etab)
&=
\left(
  1 + \sumkMl \exp(\eta_k)
\right)^{-1}
\exp(\etabt \xb)
}
$$

Comparing with
<a href='#ExpFamilyGeneralForm'>General Form of the exponential family</a>,
$$
\arrthree{
\ub(\xb) &= \xb \\
h(\xb) &= 1 \\
g(\etab) &= \left(
  1 + \sumkMl \exp(\eta_k)
\right)^{-1}
}
$$

p(x \mid \mu, \sigma^2)
&=
\dfrac{1}{(2\pi\sigma^2)^{1/2}}

}
$$


# ML and Sufficient statistics

Consider the problem of estimating the parameter vector $\eta$ in the general exponential family given by
$$
p(\xb \mid \eta)
=
h(\xb) g(\eta)
\exp \left\{ \eta^T \ub(\xb) \right\}
$$

Also
$$
g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} ~d\xb = 1
$$

Lets do it using ML. Take the gradient on both side wrt $\eta$, we have
$$
\arrthree{
\nabla g(\eta)
&\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} ~d\xb
\\
+
&g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} \ub(\xb) ~d\xb
&=0
}
$$

Since
$g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} ~d\xb = 1
$

We have,
$$
-\dfrac{1}{g(\eta)}
\nabla g(\eta)
=g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} \ub(\xb) ~d\xb
= \E{\ub(\xb)}
\\
\text{That is } \\
-\nabla \ln g(\eta) = \E{\ub(\xb)}
$$

Consider IID $\Xb = \{ \xb_1,\cdots,\xb_n \}$.
$$
p(\Xb \mid \etab)
=
\left(
  \prodnN h(\xb_n)
\right)
g(\etab)^N
\exp
\left\{
  \etabt \sumnN \ub(\xb_n)
\right\}
$$
Setting the gradient of ln p wrt $\eta$ to zero, we get
$$
N ~\nabla \ln g(\eta_{ML}) + \sumnN \ub(\xb_n) = 0 \\
-\nabla \ln g(\eta_{ML}) = \dfrac{1}{N} \sumnN \ub(\xb_n)
$$
* This can in principle be solved for $\eta_{ML}$
* Solution for ML depends on data thro' $\sum_n \ub(\xb_n)$
* This is called the sufficient statistic of the distribution.
* Need not store the entire data set just the suff. stat. will do
* Examples
  * Bernoulli: $\ub(x)$ is given by just x, so need keep the sum of {$x_n$}
  * Gaussian: $\ub(x) = (x,x^2)^T$, so keep sum of x's and their squares
* As $N \rightarrow \infty$, the RHS becomes $\E{\ub(\xb)}$, hence $\eta_{ML}$ approaches the tru value of $\eta$

# Conjugate priors

* For bernoulli, we had beta
* For gaussian
  * for mean, another Gaussian
  * for precision, Wishart distribution
* In general, for a given dist. $p(\xb \mid \etab)$, we seek a prior $p(\etab)$ so that the posterior is of the same form as the prior. If the exponential family is given by
$$
p(\xb \mid \eta)
=
h(\xb) g(\eta)
\exp \left\{ \eta^T \ub(\xb) \right\}
$$

The conjugate prior takes the form
$$
p(\etab \mid \chib, \nu)
=
f(\chib, \nu)
g(\etab)^\nu
\exp
\left\{
  \nu ~\etabt ~\chib
\right\}
$$

Here
$$
\arrthree{
f(\chib, \nu) &= \text{normalization coefficient} \\
g(\etab) &= \text{same as the one in the exp family} \\
}
$$

To se that this is indeed the conjugate, lets multiply prior by likelihood function to obtain the posterior as follows
$$
p(\etab \mid \Xb, \chib, \nu)
\propto
g(\etab)^{\nu+N}
\exp
\left\{
  \etabt
  \left(
    \sumnN \ub(\xb_n) + \nu \chib
  \right)
\right\}
$$

* This takes the same functional form as the prior, confirming conjugacy.
* $\nu$ can be interpreted as the effective number of pseudo-observations in the prior,

# Noninformative priors

* when we have little idea of what form the distribution should take.
* ~ has little influence on the posterior
* Letting the data speak for themselves
* Say we have a distribution p(x | A) governed by A.
  * we can have p(A) = const as a prior
  * If A is discrete with K states, prior prob of each state is 1/K
  * For continuous case, there are two problems
    * if domain of A is unbounded, prior cannot be correctly normalized. Such priors are called *improper* priors. These can be used provided the posterior is *proper*
    * Non linear change of variables will be PITA
    
* If function h(x) is constant, let $x = y^2 \text{ then } \hat{h}(y) = h(y^2)$
* If the density $p_x(x)$ is to be constans, the density of y is
$$
p_y(y) = p_x(x) \left| \dfrac{dx}{dy} \right|
= p_x(y^2) 2y \propto y
$$
* Thus the density over y will not be constant, while that over x will be constant.

## Translation invariant prior

if the density takes the form
$$
p(x \mid \mu) = f(x - \mu)
$$
then the parameter $\mu$ is called the location parameter.

This family of densities exhibits translation invariance coz if we shift x by a constant c to give $\hat{x} = x + c$ and $\hat{\mu} = \mu + c$, then
$$
p(\hat{x} \mid \hat{\mu}) = f(\hat{x} - \hat{\mu})
$$

Now lets choose a prior that reflects this translation invariant property.
We choose one that assigns equal probability mass to an interval $A \le \mu \le B$
as to the shifted one $A-c \le \mu \le B-c$. That is
$$
\int_{A}^{B} p(\mu) ~d\mu = \int_{A-c}^{B-c} p(\mu) ~d\mu
\\
\text{Let } \mu^{\prime} = \mu + c, \text{ which then becomes }\\
\int_{A}^{B} p(\mu) ~d\mu = \int_{A}^{B} p(\mu^{\prime}-c) ~d\mu^{\prime}
$$

Since this must hold for all chioces of A and B, we have
$$
p(\mu - c) = p(\mu)
$$
which means that $p(\mu)$ is a constant. 

### Example - Translation invariance

* An example of a location parameter would be the mean $\mu$ of a Gaussian distribution.
* The conjugate prior distribution for $\mu$ in this case is a Gaussian given by
$$
p(\mu|\mu_0, \sigma_0^2) = \Nl{\mu}{\mu_0}{\sigma_0^2}
$$
* We can obtain a noninformative prior by taking the limit $\sigma_0^2 \rightarrow \infty$

Consider the formulas obtained in Bayesian inference for Gaussian for the 
[case](3-gaussian-bayesian-inference.ipynb#Summary---case-1)
when variance is known and mean isn't.
The mean and variance were found to be

$$
\arrthree{
\mu_N &= \frac{N\sigma_0^2}{N\sigma_0^2 + \sigma^2} \mu_{ML} +
\frac{\sigma^2}{N\sigma_0^2 + \sigma^2} \mu_0
\\
\frac{1}{\sigma_N^2} &= \frac{N}{\sigma^2} + \frac{1}{\sigma_0^2}
}
$$

* From these forms, we see that, when $\sigma_0^2 \rightarrow \infty$,  this gives a posterior distribution over $\mu$ in which the contributions from the prior vanish.

## Scale invariant prior

Consider the density of the form
$$
p(x \mid \sigma) = \dfrac{1}{\sigma} f\left(\dfrac{x}{\sigma}\right), \sigma \gt 0
$$

This would be a normalized density if f(x) is correctly normalized.
$$
\int p(x \mid \sigma)
=
\dfrac{1}{\sigma}
\int_{\sigma A}^{\sigma B}
f\left(\dfrac{x}{\sigma}\right) ~dx
\\
\text{Here (A,B) form the domain of f}\\
\text{Let } y = \dfrac{x}{\sigma}, \text{ Then}
\\
p(x \mid \sigma)
=
\dfrac{1}{\sigma}
\int_{A}^{B}
f\left(y\right) ~\sigma ~dy
=
\int_{A}^{B} f(y) dy = 1
$$

The parameter $\sigma$ is called the scale parameter.
The density exhibits scale invariance coz if we scale x by c to give
$\hat{x} = cx, \hat{\sigma} = c\sigma$, we have
$$
p(\hat{x} \mid \hat{\sigma})
=
\dfrac{1}{\hat{\sigma}}
~f\left( \dfrac{\hat{x}}{\hat{\sigma}}\right)
$$

Now we have to choose a prior which reflects this scale invariance.
The prior should have equal probability mass on both $A \le \sigma \le B$
and the scaled interval $A/c \le \sigma \le B/c$. That is,
$$
\int_{A}^{B} p(\sigma) ~d\sigma
=
\int_{A/c}^{B/c} p(\sigma) ~d\sigma
=
\int_{A}^{B}
p
\left(
\dfrac{1}{c} \sigma 
\right)
\dfrac{1}{c} ~d\sigma
$$
Since this must hold for all A,B, we have
$$
p(\sigma)
=
p
\left(
\dfrac{1}{c} \sigma 
\right)
\dfrac{1}{c}
$$

Hence $p(\sigma) \propto 1/\sigma$

### Implications

The implications are quite sexy. I think this makes the integral converge.  
Rationale:  
If $\Lim{n \rightarrow \infty} a_n = 0$ and $a_n < a_{n-1}$, then the series converges

Also, the probability mass is more near the origin (similar to the log function?)

### Example -- Scale invariance

Consider
$$
\Nl{x}{\mu}{\sigmasqr}
=
\inv{\sigma} \EXP{-\left(\frac{\tilde{x}}{\sigmasqr}\right)}
$$
where $\tilde{x} = x - \mu$.

* Using the transformation rule for densities, we see that $p(\sigma) \propto 1/\sigma$
* This corresponds to a distribution over $\lambda$ of the form $p(\lambda) \propto 1/\lambda$
* we have seen that the conjugate prior for unknown variance and known mean was a Gamma distribution $\text{Gam}(\lambda \mid a_0, b_0)$
* The noninformative prior is given by the special case $a_0 = b_0 = 0$

The posterior parameters for this [case](3-gaussian-bayesian-inference.ipynb#Summary---case-2) is
$$
\arrthree{
a_N &= a_0 + \frac{N}{2}
\\
b_N &= b_0 + \frac{1}{2} \sumnN (x_n - \mu)^2
}
$$

When $a_0 = b_0 = 0$, we have
$$
\arrthree{
a_N &= \frac{N}{2} \\
b_n &= \half\sumnN (x_n - \mu)^2
}
$$
