# Student's t-distribution

Conjugate prior is a gamma distribution.
If we have a univariate gaussian and a Gamma prior,
we can integrate out the precision to obtain a marginal
distribution of x in the form

$$
\arrthree{
p(x \mid \mu,a,b)
&=
{\displaystyle \int_{0}^{\infty}}
\mathcal{N}(x \mid \mu, \tau^{-1})
~\gam(\tau \mid a,b) ~d\tau
\\
&=
\displaystyle \int_{0}^{\infty}
\dfrac{b^a e^{-b\tau}\tau^{a-1}}{\Gamma(a)}
\left( \dfrac{\tau}{2\pi} \right)^{1/2}
\exp
\left\{
  -\dfrac{\tau}{2} (x-\mu)^2
\right\}
~d\tau
\\
&=
\dfrac{b^a}{\Gamma(a)}
\left( \dfrac{1}{2\pi} \right)^{1/2}
{\displaystyle \int_{0}^{\infty}}
e^{-b\tau} \tau^{a-1} \tau^{1/2}
\exp
\left\{
  -\dfrac{\tau}{2} (x-\mu)^2
\right\}
~d\tau
\\
&=
\dfrac{b^a}{\Gamma(a)}
\left( \dfrac{1}{2\pi} \right)^{1/2}
{\displaystyle \int_{0}^{\infty}}
\tau^{a-1/2}
\exp
\left\{
  -b\tau
  -\dfrac{\tau}{2} (x-\mu)^2
\right\}
~d\tau
\\
\text{substitute }
z &= \tau
\left[
  b + \frac{(x-\mu)^2}{2}
\right]
\\
&=
\dfrac{b^a}{\Gamma(a)}
\left( \dfrac{1}{2\pi} \right)^{1/2}
\left[ b + \frac{(x - \mu)^2}{2} \right]^{-(a-1/2)-1}
{\displaystyle \int_{0}^{\infty}}
z^{a-1/2}
\exp(-z)
~dz
\\
&=
\dfrac{b^a}{\Gamma(a)}
\left( \dfrac{1}{2\pi} \right)^{1/2}
\left[ b + \frac{(x - \mu)^2}{2} \right]^{-a-1/2}
\Gamma \left(a+\dfrac{1}{2}\right)
\\
&=
\dfrac{\Gamma(a+1/2)}{\Gamma(a)}
\left( \dfrac{1}{2b\pi} \right)^{1/2}
\left[ 1 + \frac{(x - \mu)^2}{2b} \right]^{-a-1/2}
}
$$

If we set $\nu = 2a \text{ and } \lambda = a/b$, then  
$b = a/\lambda = \nu/2\lambda$

Then $p(x \mid \mu,a,b)$ takes the form
$$
\text{St}(x \mid \mu, \lambda, \nu)
=
\frac{\Gamma(\nu/2 + 1/2)}{\Gamma(\nu/2)}
\left( \frac{\lambda}{\nu \pi} \right)^{1/2}
\left[ 1 + \frac{\lambda(x-\mu)^2}{\nu}  \right]^{-\nu/2-1/2}
$$
where
* $\lambda$ is called the precision, though its inverse is not the variance.
* $\nu$: degrees of freedom
  * for $\nu=1$, it becomes Cauchy distribution
  * for $\nu \rightarrow \infty$, it becomes Gaussian with mean $\mu$ and precision $\lambda$

# Coolness

* Obtained by adding infinite number of gaussians
* Has longer tails, which makes it robust to outliers
* ML solution can be obtained using EM method
* Least squares does not exhibit robustness since it 
  corresponds to the ML under a conditional Gaussian
* By basing the regression model on a heavy-tailed distribution,
  such as the t-distribution, we obtain a robust model.

$$
\arrthree{
p(x \mid \mu,a,b)
&=
{\displaystyle \int_{0}^{\infty}}
\mathcal{N}(x \mid \mu, \tau^{-1})
~\gam(\tau \mid a,b) ~d\tau
\\
\text{substitute }
\nu = 2a & ~~~ \lambda = a/b, \eta = \tau b/a
\\
&=
{\displaystyle \int_{0}^{\infty}}
\mathcal{N}\left(x \mid \mu, (\eta\lambda)^{-1} \right)
\gam\left( \eta \mid \dfrac{\nu}{2}, \dfrac{\nu}{2} \right)
~d\eta
}
$$

In a mutivariate case, this becomes
$$
\arrthree{
\text{St}(\xb \mid \mub, \Lambda, \nu)
&=
{\displaystyle \int_{0}^{\infty}}
\mathcal{N}\left(\xb \mid \mub, (\eta\Lambda)^{-1} \right)
\gam\left( \eta \biggm | \dfrac{\nu}{2}, \dfrac{\nu}{2} \right)
~d\eta
\\
&=
\dfrac{\Gamma(D/2 + \nu/2)}{\Gamma(\nu/2)}
\dfrac{\Lambda^{1/2}}{(\nu \pi)^{D/2}}
\left[ 1 + \dfrac{\Delta^2}{\nu}  \right]^{-D/2-\nu/2}
\\
\text{where }
D &= \text{ dimensionality of } \xb \\
\Delta^2 &= (\xb-\mub)^T\Lambda(\xb-\mub)
&&\commentgray{
  $\begin{matrix}
  \text{Squared }\\ \text{Mahalanobis } \text{distance}
  \end{matrix}
  $
 }
\\
\E{\xb} &= \mub & \text{if } \nu >1 \\
\cov{\xb} &= \dfrac{\nu}{\nu-2}\li & \text{if } \nu >2 \\
\text{mode}[\xb] &= \mu
}
$$
