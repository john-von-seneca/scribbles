# IRLS
iterative reweighted least squares

* in lin regression, the ml solution has a closed form coz of gaussian noise model
  * consequence of the quadratic dependence of the log likelihood on **w**
* in log reg, there is no closed form since $\sigma$ is nonlinear
* but the error function is convex and hence has a unique min
* the error fucntion can be minimized by iterative tech based on newton-raphson iterative optimization scheme.
  * NR uses a local quadratic approximation of the objective

nr update for minimizing E(**w**) takes the form  
$ \ds \wnr{new} = \wnr{old} - \inv{\Hb} \nabla E(\wb)$  
where **H** is Hessian: second derivatives of E wrt **w**

## Newton-Raphson for lin reg

lets apply Newton-Raphson to lin reg with sum-of-squares error function
[here](../linear-models-for-regression/linear-basis-function-models.ipynb)

Remember
$$
\Phi(\mathbf{X}) =
\mat{
\phi_0(x_1) & \phi_1(x_1) & \cdots & \phi_{M-1}(x_1)\\
\phi_0(x_2) & \phi_1(x_2) & \cdots & \phi_{M-1}(x_2)\\
\vdots      & \vdots      & \ddots & \vdots\\
\phi_0(x_N) & \phi_1(x_N) & \cdots & \phi_{M-1}(x_N)\\
}
\equiv \mat{ \phi_0^T \\ \vdots \\ \phi_{N-1}^T }
\equiv \mat{
\varphi_0 & \cdots & \varphi_{M-1}
}
$$

$$
\arrthree{
E(\wb) &= \half \sumnN \Brace{t_n - \wt \phi_n}^2 \\
\nabla E(\wb) &=  \sumnN \Brace{t_n - \wt \phi_n} \phi_n
= \Phibt \Phib \wb - \Phibt \tb \\
\Hb = \nabla \nabla E(\wb) &=
\sumnN \phi_n \phi_n^T = \Phibt \Phib
}
$$

hence, the nr update becomes
$$
\arrthree{
\wnr{new} &=
\wnr{old} - \invp{\Phibt \Phib}
\Brace{\Phibt \Phib \wnr{old} - \Phibt \tb}
\\ &=
\invp{\Phibt \Phib} \Phibt \tb
}
$$

whos=a, this is the same as leat squares

## Newton Raphson for logistic regression

### Derivatives 

as derived [here](#cross-entropy-error)
$$
\arrthree{
\nabla E(\wb) &=
\sumnN \Paran{y_n - t_n} \phi_n
= \Phibt (\yb - \tb)
\\
\Hb = \nabla \nabla E(\wb) &=
\sumnN y_n (1-y_n) \phi_n \phi_n^T
= \Phibt \Rb \Phib
}
$$
where the diagonal matrix **R** is $\Rb_{nn} = y_n (1-y_n)$

### convexity 
* Now this is equivalent to an SVD formulation where **R** is the diagonal matrix of eigen values. 
* Since $0 < y_n < 1$, we have all the eigen values > 0
* Hence **H** is positive definite.
* hence the error function is a convex function of **w**
* and a unique minimum

### Newton-Raphson update

the Newton-Raphson update for log reg becomes
$$
\arrthree{
\wnr{new}
&=
\wnr{old} - \invp{\Phibt \Rb \Phib} \Phibt (\yb - \tb)
\\ &=
\invp{\Phibt \Rb \Phib} \Brace{
\Phibt \Rb \Phib \wnr{old} - \Phibt (\yb - \tb)
}
\\ &=
\invp{\Phibt \Rb \Phib} \Phibt \Brace{
\Rb \Phib \wnr{old} - (\yb - \tb)
}
\\ &=
\invp{\Phibt \Rb \Phib} \Phibt \Rb \Brace{
\Phib \wnr{old} - \inv{\Rb} (\yb - \tb)
}
\\ &=
\invp{\Phibt \Rb \Phib} \Phibt \Rb \zb
}
$$
where $\zb = \Phib \wnr{old} - \inv{\Rb} (\yb - \tb)$

### why the long name

* this takes a set of normal equations for a weighted LS
* since the weighing matrix **R** aint constant but depends on **w**
  * we need to apply normal equations iteratively
  * each time finding **w**
  * to compute a new **R**

hence this is called Iterative Reweighted Least Squares.

### what is R?

Since
$$
\arrthree{
\E{t} &= \sigma(\xb) = y
\\
\V{t} &= \E{t^2} - \E{t}^2 = \E{t} - \E{t^2} = y(1-y)
}
$$
since $t^2 = t$ for $t \in \Brace{0, 1}$

Thus, the elements of **R** are the variances

### what is z?

The quantity $z_n$ can be interpreted as an effective target value in this space obtained by making a local linear approximation
to the logistic sigmoid function
around the current operating point $\wnr{old}$. that is
$$
\arrthree{
a_n(w)
& \approx
a_n(\wnr{old}) + 
\left. \diff{a_n}{y_n} \right|_{\wnr{old}}
(t_n - y_n)
\\ &=
\phi_n^T \wnr{old} 
- \frac{y_n - t_n}{y_n (1-y_n)}
\\ &= z_n
}
$$
