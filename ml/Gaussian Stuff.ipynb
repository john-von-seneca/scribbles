# Introduction

<div id='StandardForm' \>
Standard Form  
$$
\Nstdx
=
\multivarcoeff \exp \multivarexpstd
$$

<div id='QuadraticForm'\>
Quadratic Form:
$$
\Delta^2
=
(\xb - \mub)^T
\mathbf{\Sigma}^{-1}
(\xb - \mub)
$$
$\Delta$ is called the Mahalanobis Distance

$\Sigma$
* symmetric since the non-symmetric part will disappear if written as a combination of a symmetric part + asymmetric part [Proof](/notebooks/void-main/appendix/linear-algebra.ipynb#Sum-of-symm-and-anti-symm)
* Real valued
  * Hence the eigen values are real
  * Eigen vectors are orthonormal

* Eigen: If $\left\{\u_i\right\}, \left\{\lambda_i\right\}$ are the eigenvectors and eigenvalues of $\Sigma$, then
$$
\Sigma ~ \u_i = \lambda_i \u_i
$$

* Orthonormality
$$
\u_i^T \u_j
=
\begin{cases}
1 & \text{if i == j}\\
0 & \text{otherwise}
\end{cases}
$$

* <div id='SpectralDecomposition' \>Spectral Decomposition  
$
\begin{array}{ll}
\Sigma      &= \sum_{i=1}^{D} \lambda_i \u_i \u_i^T\\
\Sigma^{-1} &= \sum_{i=1}^{D} \frac{1}{\lambda_i} \u_i \u_i^T
\end{array}
$

* Determinant
$$
\left\lvert
\Sigma
\right\rvert
^{1/2}
=
\prod_{i=1}^{D} \lambda_i^{1/2}
$$

Multivariate to Independent Univariates
=====================

Substituting <a href="#SpectralDecomposition">Spectral Decomposition</a> into <a href="#QuadraticForm">Quadratic Form</a>, we get
\begin{array}{llr}
\Delta^2
&=
(\xb - \mub)^T
\left(
  \sum_{i=1}^{D} \frac{1}{\lambda_i} \u_i \u_i^T
\right)
(\xb - \mub)
\\
&=
\sum_{i=1}^{D} \frac{y_i^2}{\lambda_i}
\end{array}
where $y_i = \u_i^T (\xb - \mub)$. Further,
\begin{array}{ll}
\text{If } \yb
= 
\left[
\begin{matrix}
y_1 \\ \vdots \\ y_D
\end{matrix}
\right] 
&
\text{Then }
\yb = \Ub (\xb - \mub)
\\
\text{where }
\Ub
&=
\mat{\u_1^T \\ \vdots \\ \u_D^T}
\end{array}

We know that $p(x) = p(y) \left| \frac{dx}{dy} \right|$.  
Now
$$
\Jb_{ij} = \Partial{x_i}{y_j} = \Ub_{ji}
$$

$\Mod{\Jb}^2 = \Mod{\Ut}^2 = \Mod{\Ut}\Mod{\Ub} = \Mod{\Ut \Ub} = \Mod{\I} = 1$.  
Hence $\Mod{\Jb} = 1$

Hence, in the $y_j$ coordinate system, <a href="#StandardForm">Standard Form</a> becomes
$$
p(\yb)
=
p(\xb)
\lvert \mathbf{J} \rvert
=
\prod_{i=1}^{D}
\frac{1}{(2\pi\lambda_i)^{1/2}}
\exp \left(-\frac{y_i^2}{2\lambda_i}\right)
$$
Thus, eigenvectors definea new set of shifted and rotated coordinate system,
where it becomes a set of independent univariate Gaussians with 
$y_i = \u_i^T (\xb - \mu), \lambda_i$ as mean and covariance

Moments
====

Mean
----
\begin{array}{llr}
\E{\xb}
&=
\multivarcoeff
\int \exp \multivarexpstd \xb ~dx
\\
&=
\multivarcoeff
\int \exp \multivarexpx{\zb} ~(\zb + \mub) ~d\zb
&
\commentgray{[1]}
\\
&=
\mub
\multivarcoeff
\int \exp \multivarexpx{\zb} ~d\zb
&
\commentgray{[2]}
\\
\implies
\E{\xb}
&=
\mub
\end{array}

1. Sub $\zb = (\xb - \mub)$
1. Since the exponent is a even component and the integral ranges from $(-\infty, \infty)$,  
$\zb$ in the factor $(\zb + \mub)$ will get the fuck lost, leaving just $\mub$.


Second Moment
------------

\begin{array}{llr}
\E{\xb \xb^T}
&=
\multivarcoeff
\int \exp \multivarexpstd \xb \xb^T ~d\xb
\\
&=
\multivarcoeff
\int \exp \multivarexpx{\zb} (\zb+\mub)(\zb+\mub)^T ~d\zb
&
\commentgray{[1]}
\\
&=
\multivarcoeff
\int \exp \multivarexpx{\zb} (\zb \zb^T + \mub \mub^T) ~d\zb
&
\commentgray{[2]}
\\
&=
\mub \mub^T
+
\multivarcoeff
\int \exp \multivarexpx{\zb} \zb \zb^T ~d\zb
\\
&=
\mub \mub^T
+
\multivarcoeff
\sum_{i=1}^{D} \sum_{j=1}^{D}
\u_i \u_j
\int \exp
\multivarexpx{\zb} \zb \zb^T
y_i y_j ~d\yb
&
\commentgray{[3]}
\\
&=
\mub \mub^T
+
\multivarcoeff
\sum_{i=1}^{D} \sum_{j=1}^{D}
\u_i \u_j
\int \exp
\left\{
  -\sum_{k=1}^{D}
     \frac{y_k^2}
          {2\lambda_k}
\right\}
y_i y_j ~d\yb
&
\commentgray{[4]}
\\
&=
\mub \mub^T
+
\multivarcoeff
\sum_{i=1}^{D}
\u_i \u_i^T
\int \exp
\left\{
  -\sum_{k=1}^{D}
     \frac{y_k^2}
          {2\lambda_k}
\right\}
y_i^2 ~d\yb
&
\commentgray{[5]}
\\
&=
\mub \mub^T
+
\sum_{i=1}^{D} \u_i \u_i^T \lambda_i
&
\commentgray{[6]}
\\
\E{\xb \xb^T}
&=
\mub \mub^T
+
\Sigma
&
\commentgray{[7]}
\end{array}

1. Sub $\zb = (\xb - \mub)$
2. Exponent is even $\implies \zb$ terms will vanish
3. $\zb = \sum_{i=1}^{D} y_i \u_j$
4. <a href='#SpectralDecomposition'>SD</a> of $\Sigma^{-1}$:
   $\left( \sum_{i=1}^{D} y_i \u_j^T \right)
    \left( \sum_{i=1}^{D} \frac{1}{\lambda_i} \u_i \u_i^T \right)
    \left( \sum_{i=1}^{D} y_i \u_j \right)
   $
5. By Symmetry,
$$
\text{integral term containing } y_i y_j = 
\begin{cases}
  y_i^2 & \text{if i==j}\\
  0 & \text{if i}\ne\text{j}
\end{cases}
$$
6. $\E{x^2} = \mu^2 + \sigma^2 = 0 + (\sqrt{\lambda_i})^2 = \lambda_i$
7. <a href='#SpectralDecomposition'>SD</a> of $\Sigma = \sum_{i=1}^{D} \lambda_i \u_i \u_i^T$

Covariance
---------
\begin{array}{llr}
\cov{\xb}
&=
\E{\left(\xb - \E{\xb}\right)\left(\xb - \E{\xb}\right)^T}
\\
&=
\E{\left(\xb - \mub\right)\left(\xb - \mub\right)^T}
&
\commentgray{$\E{\xb} = \mub$}
\\
&=
\E{\xb \xb^T} - \mub \mub^T
&
\commentgray{expanding}
\\
&=
\Sigma
&
\commentgray{$\E{\xb\xb^T} = \mub \mub^T + \Sigma$}
\end{array}
Hence the parameter $\Sigma$ is called the fucking covariance matrix.

Limitations
======

1. Complexity
  1. Number of parameters =
     Parameters in
     $\mu$ + Parameters in $\Sigma = D + \frac{D(D+1)}{2} = \frac{D(D+3)}{2}$ 
  1. Inverting Large $\Sigma$ becomes prohibitive.
    * can use Diagonal covariance or better isotropic covariance
    * Such a feat, although makes things computationally easy, restricts the ability to capture interesting information from data
1. Always Unimodal

Cure
----
1. Continuous Latent Variables: Tractable no. of parameters
  1. MRF's
  1. Linear dynamical systems
1. Mixture of Gaussians: Multimodal

Conditional Gaussian Distributions
===================

Let 
\begin{array}{rlr}
\xb
&=
\mat{\xab \\ \xbb}
\\
\mub
&=
\mat{\muab \\ \mubb}
\\
\Sigma
&=
\mat{\saa & \sab \\ \sba & \sbb}
\\
\text{since} \Sigma^T = \Sigma,
&
\text{we have } \saa \text{ and } \sbb \text{ being symmetric}
&\text{ and } \sba = \sab^T
\\
\Lambda &\equiv \Sigma^{-1}
\\
\Lambda
&=
\mat{\laa & \lab \\ \lba & \lbb}
\end{array}


To find the conditional $p(\xab \mid \xbb)$, we can set $\xbb$ in the joint distribution $p(\xab, \xbb)$. But there is another way. By using  <a href="#QuadraticForm">Quadratic Form</a>, 

<div id='StandardExpansion' />
Given the quadratic form, we can find the mean and covariance by completing the square. Expanding the quadratic form, we get
$$
-\frac{1}{2} (\xb-\mub)^T \Sigma^{-1} (\xb-\mub)
=
-\frac{1}{2} \xb^T \Sigma^{-1} \xb
+\xb^T \Sigma^{-1} \mub
+ const
$$
If we can equate the coefficients of any quadratic form similar to the RHS of the above equation, it becomes straightforward to find the mean and covariance.

<div id='partitionedexpansion'/>
\begin{array}{rlr}
-\frac{1}{2} (\xb - \mub)^T \Sigma^{-1} (\xb - \mub)
&=
\mat{\xab-\muab & \xbb-\mubb}
\mat{\saa & \sab \\ \sba & \sbb}
\mat{\xab-\muab \\ \xbb-\mubb}
\\
&=
-\frac{1}{2} (\xab - \muab)^T \laa (\xab - \muab)
\\
&
-\frac{1}{2} (\xab - \muab)^T \lab (\xbb - \mubb)
\\
&
-\frac{1}{2} (\xbb - \mubb)^T \lba (\xab - \muab)
\\
&
-\frac{1}{2} (\xbb - \mubb)^T \lbb (\xbb - \mubb)
\end{array}

Now we need to find $\mu_{\ab \mid \bb}$ and $\Sigma_{\ab \mid \bb}$. If we consider $\xbb$ as constant and pick the terms of second order in $\xab$, we get

\begin{array}{llr}
-\frac{1}{2} \xab^T \laa \xab
\end{array}

Hence, $\Sigma_{\ab \mid \bb} = \laa^{-1}$

Terms linear in $\xab$  
$$
\arrthree{
&\laa\muab
\\
&-\frac{1}{2} \lab (\xbb - \mubb)
\\
&-\frac{1}{2} \lab (\xbb - \mubb)
\\
\implies
&
\xab^T
\left\{
  \laa\muab - \lab (\xbb - \mubb)
\right\}
}
\\
\implies
\Sigma^{-1}_{\ab \mid \bb} \mu_{\ab \mid \bb}
=
\laa\muab - \lab (\xbb - \mubb)
\\
\laa \mu_{\ab \mid \bb}
=
\laa\muab - \lab (\xbb - \mubb)
\\
\mu_{\ab \mid \bb}
=
\muab - \laa^{-1} \lab (\xbb - \mubb)
\\
$$

<div id='ConditionalMoments' />
Hence, the mean and covariance of the condional are
$$
\arrthree{
\E{\mub_{\ab \mid \bb}}
&=
\muab - \laa^{-1} \lab (\xbb - \mubb)
\\
\cov{\ab \mid \bb}
&=
\laai
}
$$

<div id='ParitionInverse'/>
Since
\begin{array}{rlr}
\mat{
\Ab & \Bb\\
\Cb & \Db
}^{-1}
&=
\mat{
\Mb & -\Mb\Bb\Db^{-1}
\\
-\Db^{-1}\Cb\Mb &
\Db^{-1}+\Db^{-1}\Cb\Mb\Bb\Db^{-1}
}
\\
\text{where }
\Mb
&=
\left(
  \Ab - \Bb\Db^{-1}\Cb
\right)^{-1}
\\
\end{array}

Using the covariance matrix instead of the precision matrix, we get

$$
\arrthree{
\mat{
\saa & \sab \\
\sba & \sbb
}^{-1}
&=
\mat{\laa & \lab \\ \lba & \lbb}
\\
\laa
&=
\mathbf{M} = (\saa - \sab \sbb^{-1} \sba)^{-1}
\\
\lab
&= -\mathbf{M} \sab \sbb^{-1}
 = -(\saa - \sab \sbb^{-1} \sba)^{-1} \sab \sbb^{-1}
\\
\text{Hence}
\\
\mu_{\ab \mid \bb}
&=
\muab - \laa^{-1} \lab (\xbb - \mubb)
\\
&=
\muab
-
\laa^{-1}
(-\laa \sab \sbb^{-1})
(\xbb - \mubb)
\\
&=
\muab + \sab \sbb^{-1} (\xbb -\mubb)
\\
\Sigma_{\ab \mid \bb}
&=
\saa - \sab\sbb^{-1}\sba
}
$$

Marginal Gaussian Distributions
=================

Need to find $p(\xab) = \int p(\xab, \xbb) d\xbb$.
From <a href="#partitionedexpansion">Partitioned Expansion</a>, we have the following terms involving only $\xbb$.
$$
\arrthree{
\text{Terms having }\xbb
&=
-\frac{1}{2}(\xab-\muab)^T \lab \xbb
\\
&~~~ 
-\frac{1}{2}\xbb^T \lba (\xab-\muab)
\\
&~~~
-\frac{1}{2} \xbb^T \lbb \xbb
+ \xbb^T \lbb \mubb
\\
&=
-\frac{1}{2} \xbb^T \lbb \xbb
+ \xbb^T \lbb \mubb
- \xbb^T \lba (\xab-\muab)
\\
&=
-\frac{1}{2} \xbb^T \lbb \xbb
+ \xbb^T ( \lbb \mubb - \lba (\xab-\muab))
\\
&=
-\frac{1}{2} \xbb^T \lbb \xbb
+ \xbb^T \mb
\\
&=
-\frac{1}{2} \xbb^T \lbb \xbb
+\frac{1}{2} 2 \xbb^T \lbb \lbb^{-1} \mb
-\frac{1}{2} \mb^T \lbb^{-1} \mb
+\frac{1}{2} \mb^T \lbb^{-1} \mb
\\
&=
-\frac{1}{2}
(\xbb - \lbb^{-1} \mb)^{-1}
\lbb
(\xbb - \lbb^{-1} \mb)
+ \frac{1}{2} \mb^T \lbb^{-1} \mb
}
$$

Only the last term is dependent on $\xab$. Taking this term along with the other terms that depend on $\xab$ in <a href="#partitionedexpansion">Partitioned Expansion</a>, we get
$$
\arrthree{
\text{Terms having }\xab
&=
\frac{1}{2} \mb^T \lbbi \mb
\\ &
~~~
-\frac{1}{2}\xabt \laa \xab
+\xabt (\laa \muab + \lab \mubb)
\\
&=
( \lbb \mubb - \lba (\xab-\muab))^T \lbbi ( \lbb \mubb - \lba (\xab-\muab))
\\
&
-\frac{1}{2}\xabt \laa \xab
+\xabt (\laa \muab + \lab \mubb)
\\
&=
-\frac{1}{2} \xabt \left( 
\laa - \lab \lbbi \lba
\right) \xab
\\
& ~~~
+ \xabt \left( \laa\muab - \lab \lbbi \lba \right)^{-1} \muab
+ const
}
$$

From <a href='#StandardExpansion'>Standard Expansion</a>, we get
$$
\arrthree{
\text{Covariance} \\
\Sigma_a
&=
\left( \laa\muab - \lab \lbbi \lba \right)^{-1}
\\
\text{Mean}
&=
\Sigma^{-1} (\text{coeff of linear term})
\\
&=
\Sigma_a \left( \laa\muab - \lab \lbbi \lba \right) \muab
\\
&= \muab
}
$$

Another vantage point
$$
\text{Since }
\mat{\laa & \lab \\ \lba & \lbb}^{-1}
=
\mat{\saa & \sab \\ \sba & \sbb}
$$
we can use <a href='#ParitionInverse'>General Partition Inverse</a>,
to have $\saa = \left( \laa\muab - \lab \lbbi \lba \right)^{-1}$

<div id='MarginalMoments'/>
Thus, the mean and covariance for marginals is given by
$$
\arrthree{
\E{\xab} &= \muab \\
\cov{\xab} &= \saa
}$$

In conclusion, for partitioned gaussians, we have

\begin{array}{rllrl}
\xb
&=
\mat{\xab \\ \xbb}
&&
\mub
&=
\mat{\muab \\ \mubb}
\\
\Sigma
&=
\mat{\saa & \sab \\ \sba & \sbb}
&&
\Lambda
&=
\mat{\laa & \lab \\ \lba & \lbb}
\\
\end{array}
\begin{array}{rlr}
\text{Conditional Distribution}\\
p(\xab \mid \xbb)
&=
\mathcal{N}
\left(
  \xb \mid \mu_{\ab \mid \bb}, \laai
\right)
\\
\mu_{\ab \mid \bb}
&=
\muab - \laa^{-1} \lab (\xbb - \mubb)
\\
\text{Marginal Distribution} \\
p(\xab)
&=
\mathcal{N} \left( \xab \mid \muab, \saa \right)
\end{array}




Bayes' Theorem for Gaussian Variables
=====================

Given
* $p(\xb)$
* $p(\yb \mid \xb)$
  * mean: linear function of **x**

Maximum Likelihood for Gaussian
==================

Given a data set $\Xb = \left(\xb_1, \cdots, \xb_N \right)^T$ and each
$xb_n$ be drawn independently from a Multivariate Gaussian given by
<a href='#StandardForm'>Standard Form </a>.
The Log Likelihood is then given by
$$
\ln p(\Xb \mid \mub, \Sigma)
=
-\frac{ND}{2} \ln(2\pi)
-\frac{N}{2} \ln \left|\Sigma\right|
-\frac{1}{2} \sum_{n=1}^{N} (\xb_n - \mub)^T \Sigma^{-1} (\xb_n - \mub)
$$

The derivative of the log likelihood is given by
$$
\frac{\partial}{\partial \mub} \ln p(\Xb \mid \mub, \Sigma)
=
\sum_{n=1}^{N} \Sigma^{-1} (\xb_n - \mub)
$$
Hence, the ML estimate of the mean is
$$
\mub_{ML} = \frac{1}{N} \sum_{n=1}^{N} \xb_n
$$
The covariance is given by
$$
\Sigma_{ML} = \frac{1}{N} \sum_{n=1}^{N} (\xb_n - \mub)(\xb_n - \mub)^T
$$

Also, the expected values of these estimates are
$$
\arrthree{
\E{\mub_{ML}}
&=
\mub
\\
\E{\Sigma_{ML}}
&=
\frac{N-1}{N} \Sigma
}
$$

Bayesian Inference for Gaussian

Unknown Mean, Known Variance
------------------------------

ML framework gives point estimates for $\mub$ and $\Sigma$.
Say there is a single Gaussian variable x.
Say, the variance $\sigma^2$ is known.
Task: Infer mean $\mu$ from N observations $\Xb = \{x_1,\cdots,x_N\}$

$$
p(\Xb \mid \mu)
= \prod_{n=1}^{N} p(x_n \mid \mu)
= \frac{1}{\left(2\pi\sigma^2\right)^{N/2}}
\exp\left\{
  -\frac{1}{2\sigma^2} \sum_{n=1}^{N} (x_n-\mu)^2
\right\}
$$
This quantity is not normalized.

The likehood has exponential of a quadratic form over $\mu$.
If we choose the prior to be Gaussian, it will be a conjugate dist. to the likelihood.
$$
\arrthree{
\text{Prior }
p(\mu)
&=
\mathcal{N}\left( \mu \mid \mu_0, \sigma_0^2 \right)
\\
\text{Posterior }
p(\mu \mid \Xb)
&\propto
p(\Xb \mid \mu) p(\mu)
\\
&=
\arrthree{
-\frac{1}{2\sigma^2} \sumnN (x_n - \mu)^2
-\frac{1}{2\sigma_o^2} (\mu - \mu_0)^2
}
}
$$

We can use <a href='#StandardExpansion'>Standard Expansion</a> to find the mean and covar.

Take out the quadratic terms out in $\mu$
$$
-\frac{1}{2\sigma^2}\sumnN \mu^2
-\frac{1}{2\sigma_0^2} \mu^2
=
-\frac{N}{2\sigma^2} \mu^2
-\frac{1}{2\sigma_0^2} \mu^2
=
- \frac{1}{2}
  \left(
    \frac{N}{\sigma^2} + 
    \frac{1}{\sigma_0^2}
  \right) \mu^2
$$
Hence
$$
\color{green}{
\frac{1}{\sigma_N^2} =\frac{N}{\sigma^2} + \frac{1}{\sigma_0^2}
}
$$

Taking out the linear terms in $\mu$, we have  
$$
-\frac{1}{\sigma^2} (-\mu) \sumnN x_n
-\frac{1}{\sigma_0^2} (-\mu_0) \mu 
=\left( 
  \frac{1}{\sigma^2} N \mu_{ML} +
  \frac{1}{\sigma_0^2} \mu_0
 \right) \mu
$$
where
$$
\mu_{ML} = \frac{1}{N} \sumnN x_n
$$
Hence
$$
\arrthree{
\mu_N
=
\Sigma \times \text{coefficient}
&= 
\sigma_N^2 \times \left( 
  \frac{1}{\sigma^2} N \mu_{ML} +
  \frac{1}{\sigma_0^2} \mu_0
 \right)
\\
&=
\frac{ 
  \left( 
    \frac{1}{\sigma^2} N \mu_{ML} +
    \frac{1}{\sigma_0^2} \mu_0
  \right)
}
{
 \frac{N}{\sigma^2} + \frac{1}{\sigma_0^2}
}
\\
&=
\frac{N\sigma_0^2}{N\sigma_0^2 + \sigma^2} \mu_{ML} +
\frac{\sigma^2}{N\sigma_0^2 + \sigma^2} \mu_0
}
$$

Hence

$$
\color{green}{
\mu_N = \frac{N\sigma_0^2}{N\sigma_0^2 + \sigma^2} \mu_{ML} +
\frac{\sigma^2}{N\sigma_0^2 + \sigma^2} \mu_0
}
$$

-----------------------------

Its more convenient to work with the precision $\lambda \equiv 1 / \sigma^2$

The likelihood function becomes
$$
\arrthree{
p(\Xb \mid \lambda)
& \propto
\prodnN
\mathcal{N}(x_n \mid \mu, \lambda^{-1})
\\
&=
\lambda^{N/2}
\exp
\left\{
  -\frac{\lambda}{2}
  \sumnN (x_n - \mu)^2
\right\}
}
$$

The conjugate prior should be proportional to the product of power of $\lambda$ and the exponential of a linear function of $\lambda$. Seems we have one distribution like that or it was made for that. who knows. 

The gamma distribution is defined by
$$
\gam(\lambda \mid a,b)
=
\frac{1}{\Gamma(a)}
b^a \lambda^{a-1} \exp(-b \lambda)
$$
The mean and variance are
$$
\arrthree{
\E{\lambda} &= \frac{a}{b}\\
\V{\lambda} &= \frac{a}{b^2}\\
}
$$

Now the posterior is given by
$$
\arrthree{
p(\lambda \mid \Xb)
& \propto
\lambda^{a_0 - 1}
\exp
\left\{
  -b_0 \lambda
  -\frac{\lambda}{2} \sumnN (x_n - \mu)^2
\right\}
\\
\text{where }
a_N
&=
a_0 + \frac{N}{2}
\\
b_N
&=
b_0 + \frac{1}{2} \sumnN (x_n - \mu)^2
=
b_0 + \frac{N}{2} \sigma^{2}_{ML}
\\
\text{That is the posterior }
p(\lambda \mid \Xb)
& \propto
\gam(\lambda \mid a_N, b_N)
}
$$

Effect of observing N data points
* coefficient a
  * increase by N/2
  * $2a_0$ prior observations
* coefficient b
  * $N\sigma_{ML}^{2}/2$ contribution
  * $2a_0$ observations each having variance $2b_0/(2a_0) = b_0/a_0$

Exponential family
==================
<div id='ExpFamilyGeneralForm'/>
General form:
$$
p(\xb \mid \eta)
=
h(\xb) g(\eta)
\exp \left\{ \eta^T \ub(\xb) \right\}
$$
Here
* $\eta$ are the natural parameters of the distribution.
* $g(\eta)$ is the normalizing factor

Ex: Bernoulli
---------------
$$
\arrthree{
p(x \mid \mu)
&=
\mu^x (1-\mu)^{1-x}
\\
&=
\exp
\left\{
  x     \ln \mu +
  (1-x) \ln (1-\mu)
\right\}
\\
&=
(1-\mu)
\exp
\left\{
  \ln \left(\dfrac{\mu}{1-\mu}\right)x
\right\}
}
$$
Compare this with the general form of exponential familty, we have
$$\nu = \ln \left( \dfrac{\mu}{1-\mu} \right)$$
We obtain $\mu = \sigma(n)$ which is called logistic sigmoid function,
given by
$$
\sigma(\eta)
=
\dfrac{1}{1+\exp(-\eta)}
\\
\text{where }
\sigma(-\eta) = 1 - \sigma(\eta)
$$
Thus we can write Bernoulli as 
$$
\arrthree{
p(x \mid \mu)
&=
(1 - \sigma(\eta)) \exp(\eta x)
\\
&=
\sigma(-\eta) \exp(\eta x)
\\
\text{Thus} \\
u(x) &= x \\
h(x) &= 1 \\
g(\eta) &= \sigma(-\eta)
}
$$

Ex: Multinomial
--------------

Multinomial, for a single observation, $\xb$ takes the form
$$
\arrthree{
p(\xb \mid \mub)
&= \prod_{k=1}^{M} \mu_{k}^{x_k}
\\
&= \exp
\left\{
  \sum_{k=1}^{M} x_k \ln \mu_k
\right\}
\\
&= \exp(\etabt \xb)
& \commentgray{$\eta_k = \ln \mu_k$}
}
$$
Comparing with <a href='#ExpFamilyGeneralForm'>General Form of the exponential family</a>, we get
$$
\arrthree{
\ub(\xb) &= \xb \\
h(\xb) &= 1 \\
g(\etab) &= 1\\
}
$$

Since $\sum_{k=1}^{M} \mu_k = 1$, the #parameters is M-1. Thus
$$
\arrthree{
p(\xb \mid \mub)
&=
\exp \left\{ \sumkM x_k \ln \mu_k \right\}
\\
&=
\exp
\left\{
  \sumkMl x_k \ln \mu_k
  + \left( 1 - \sumkMl x_k \right)
    \ln \left( 1 - \sumkMl \mu_k \right)
\right\}
\\
&=
\exp
\left\{
  \sumkMl x_k \ln \left( \dfrac{\mu_k}{1-\sumjMl \mu_j} \right)
  - \ln \left( 1 - \sumkMl \mu_k \right)
\right\}
}
$$

$$
\arrthree{
\ln \left( \dfrac{\mu_k}{1-\sum_j \mu_j} \right)
&= \eta_k
\\
\dfrac{\mu_k}{1-\sum_j \mu_j} &= e^{\eta_k}
\\
\text{sum from k=1} & \text{to M-1 both sides}\\
\dfrac{\sum_k \mu_k}{1-\sum_j \mu_j} &= \sum_k e^{\eta_k}
\\
\sum_k \mu_k &= \dfrac{\sum_j e^{\eta_k}}{1 + \sum_j e^{\eta_k}}
\\
\text{Hence}
\dfrac{\mu_k}{1-\dfrac{\sum_j e^{\eta_k}}{1 + \sum_j e^{\eta_j}}} &= e^{\eta_k}
\\
\mu_k &= \dfrac{e^{\eta_k}}{1 + \sum_j e^{\eta_j}}
}
$$
THis is called the softmax function or the normalized exponential. Thus,


$$
\arrthree{
p(\xb \mid \etab)
&=
\left(
  1 + \sumkMl \exp(\eta_k)
\right)^{-1}
\exp(\etabt \xb)
}
$$

Comparing with
<a href='#ExpFamilyGeneralForm'>General Form of the exponential family</a>,
$$
\arrthree{
\ub(\xb) &= \xb \\
h(\xb) &= 1 \\
g(\etab) &= \left(
  1 + \sumkMl \exp(\eta_k)
\right)^{-1}
}
$$

ML and Sufficient statistics
----------------------------

Consider the problem of estimating the parameter vector $\eta$ in the general exponential family given by
$$
p(\xb \mid \eta)
=
h(\xb) g(\eta)
\exp \left\{ \eta^T \ub(\xb) \right\}
$$

Lets do it using ML. Take the gradient on both side wrt $\eta$, we have
$$
\arrthree{
\nabla g(\eta)
&\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} ~d\xb
\\
+
&g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} \ub(\xb) ~d\xb
&=0
}
$$

Since
$g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} ~d\xb = 1
$

We have,
$$
-\dfrac{1}{g(\eta)}
\nabla g(\eta)
=g(\eta)
\int h(\xb) \exp \left\{ \eta^T \ub(\xb) \right\} \ub(\xb) ~d\xb
= \E{\ub(\xb)}
\\
\text{That is } \\
-\nabla \ln g(\eta) = \E{\ub(\xb)}
$$



Consider IID $\Xb = \{ \xb_1,\cdots,\xb_n \}$.
$$
p(\Xb \mid \etab)
=
\left(
  \prodnN h(\xb_n)
\right)
g(\etab)^N
\exp
\left\{
  \etabt \sumnN \ub(\xb_n)
\right\}
$$
Setting the gradient of ln p wrt $\eta$ to zero, we get
$$
N ~\nabla \ln g(\eta_{ML}) + \sumnN \ub(\xb_n) = 0 \\
-\nabla \ln g(\eta_{ML}) = \dfrac{1}{N} \sumnN \ub(\xb_n)
$$
* This can in principle be solved for $\eta_{ML}$
* Solution for ML depends on data thro' $\sum_n \ub(\xb_n)$
* This is called the sufficient statistic of the distribution.
* Need not store the entire data set just the suff. stat. will do
* Examples
  * Bernoulli: $\ub(x)$ is given by just x, so need keep the sum of {$x_n$}
  * Gaussian: $\ub(x) = (x,x^2)^T$, so keep sum of x's and their squares
* As $N \rightarrow \infty$, the RHS becomes $\E{\ub(\xb)}$, hence $\eta_{ML}$ approaches the tru value of $\eta$

Conjugate priors
----------------
* For bernoulli, we had beta
* For gaussian
  * for mean, another Gaussian
  * for precision, Wishart distribution
* In general, for a given dist. $p(\xb \mid \etab)$, we seek a prior $p(\etab)$ so that the posterior is of the same form as the prior. If the exponential family is given by
$$
p(\xb \mid \eta)
=
h(\xb) g(\eta)
\exp \left\{ \eta^T \ub(\xb) \right\}
$$

The conjugate prior takes the form
$$
p(\etab \mid \chib, \nu)
=
f(\chib, \nu)
g(\etab)^\nu
\exp
\left\{
  \nu ~\etabt ~\chib
\right\}
$$

Here
$$
\arrthree{
f(\chib, \nu) &= \text{normalization coefficient} \\
g(\etab) &= \text{same as the one in the exp family} \\
}
$$

To se that this is indeed the conjugate, lets multiply prior by likelihood function to obtain the posterior as follows
$$
p(\etab \mid \Xb, \chib, \nu)
\propto
g(\etab)^{\nu+N}
\exp
\left\{
  \etabt
  \left(
    \sumnN \ub(\xb_n) + \nu \chib
  \right)
\right\}
$$
This takes the same functional form as the prior, confirming conjugacy.
$\nu$ can be interpreted as the effective number of pseudo-observations in the prior,

**Example: Translation invariant prior**

if the density takes the form
$$
p(x \mid \mu) = f(x - \mu)
$$
then the parameter $\mu$ is called the location parameter.

This family of densities exhibits translation invariance coz if we shift x by a constant c to give $\hat{x} = x + c$ and $\hat{\mu} = \mu + c$, then
$$
p(\hat{x} \mid \hat{\mu}) = f(\hat{x} - \hat{\mu})
$$

Now lets choose a prior that reflects this translation invariant property.
We choose one that assigns equal probability mass to an interval $A \le \mu \le B$
as to the shifted one $A-c \le \mu \le B-c$. That is
$$
\int_{A}^{B} p(\mu) ~d\mu = \int_{A-c}^{B-c} p(\mu) ~d\mu
\\
\text{Let } \mu^{\prime} = \mu + c, \text{ which then becomes }\\
\int_{A}^{B} p(\mu) ~d\mu = \int_{A}^{B} p(\mu^{\prime}-c) ~d\mu^{\prime}
$$
Since this must hold for all chioces of A and B, we have
$$
p(\mu - c) = p(\mu)
$$
which means that $p(\mu)$ is a constant. 

**Example: Scale invariant prior**

Consider the density of the form
$$
p(x \mid \sigma) = \dfrac{1}{\sigma} f\left(\dfrac{x}{\sigma}\right), \sigma \gt 0
$$

This would be a normalized density if f(x) is correctly normalized.
$$
\int p(x \mid \sigma)
=
\dfrac{1}{\sigma}
\int_{\sigma A}^{\sigma B}
f\left(\dfrac{x}{\sigma}\right) ~dx
\\
\text{Here (A,B) form the domain of f}\\
\text{Let } y = \dfrac{x}{\sigma}, \text{ Then}
\\
p(x \mid \sigma)
=
\dfrac{1}{\sigma}
\int_{A}^{B}
f\left(y\right) ~\sigma ~dy
=
\int_{A}^{B} f(y) dy = 1
$$

The parameter $\sigma$ is called the scale parameter.
The density exhibits scale invariance coz if we scale x by c to give
$\hat{x} = cx, \hat{\sigma} = c\sigma$, we have
$$
p(\hat{x} \mid \hat{\sigma})
=
\dfrac{1}{\hat{\sigma}}
~f\left( \dfrac{\hat{x}}{\hat{\sigma}}\right)
$$

Now we have to choose a prior which reflects this scale invariance.
The prior should have equal probability mass on both $A \le \sigma \le B$
and the scaled interval $A/c \le \sigma \le B/c$. That is,
$$
\int_{A}^{B} p(\sigma) ~d\sigma
=
\int_{A/c}^{B/c} p(\sigma) ~d\sigma
=
\int_{A}^{B}
p
\left(
\dfrac{1}{c} \sigma 
\right)
\dfrac{1}{c} ~d\sigma
$$
Since this must hold for all A,B, we have
$$
p(\sigma)
=
p
\left(
\dfrac{1}{c} \sigma 
\right)
\dfrac{1}{c}
$$

Hence $p(\sigma) \propto 1/\sigma$
