```python
class MaxHeap:
    def __init__(self, n):
        self.n = n
        self.head = 0
        self.array = [None for _ in range(2*n)]
    def remove(self, value):
        ix = self.find(value)
        if ix is None: raise('value not found')
        self.array[ix] = -1
        self.reverse_heapify(ix)
        self.update_head()
    def insert(self, value):
        #print('insert', value)
        self.array[self.head] = value
        self.heapify(self.head)
        self.update_head()
        #print(self.array)
    def heapify(self, ix):
        parent, sibling = self.family(ix)
        if parent is None: return
        if self.array[parent] < self.array[ix]:
            self.array[parent],self.array[ix] = self.array[ix], self.array[parent]
            self.heapify(parent)
    def reverse_heapify(self, ix):
        ix_l, ix_r = self.child_left(ix), self.child_right(ix)
        l = self.array[ix_l] if ix_l and ix_l<len(self.array) else None
        r = self.array[ix_r] if ix_r and ix_r<len(self.array) else None
        max_ix = 'parent'
        m = self.array[ix]
        if l is not None and l > m:
            m, max_ix = l, 'left'
        if r is not None and r > m:
            m, max_ix = r, 'right'

        print('ix', ix, ':', self.array[ix] , 
              'c:', [ix_l, ix_r], ':', [l, r], 
              'max', max_ix, ':', m)
        if max_ix == 'left':
            self.array[ix], self.array[ix_l] = self.array[ix_l], self.array[ix]
            self.reverse_heapify(ix_l)
        elif max_ix== 'right':
            self.array[ix], self.array[ix_r] = self.array[ix_r], self.array[ix]
            self.reverse_heapify(ix_r)
        else:
            self.array[ix] = None
    def find(self, value):
        for ix in range(len(self.array)):
            if self.array[ix] == value:
                return ix
        return None
    def family(self,ix):
        return [self.parent(ix), self.sibling(ix)]
    def parent(self, ix):
        ix_parent = math.floor((ix-1)/2.0)
        return ix_parent if ix_parent>=0 else None
    def sibling(self, ix):
        ix_parent = self.parent(ix)
        if ix_parent is None: return None
        return ix+1 if ix==self.child_left(ix_parent) else ix-1
    def child_left(self, ix):
        ix_lc = 2*ix + 1
        return ix_lc if ix_lc < len(self.array) else None
    def child_right(self, ix):
        ix_rc = 2*ix + 2
        return ix_rc if ix_rc < len(self.array) else None
    def update_head(self): 
        for ix in range(len(self.array)):
            if self.array[ix] is None:
                self.head = ix
                break
    def graphviz(self):
        g = UGraph()
        for ix in range(len(self.array)):
            if self.array[ix] is None: continue
            l, r = self.child_left(ix), self.child_right(ix)
            if l is not None and self.array[l] is not None:
                g.edge(str(self.array[ix]), str(self.array[l]))
            if r is not None  and self.array[r] is not None:
                g.edge(str(self.array[ix]), str(self.array[r]))
        return g
```

```python
# x-start, x-end, height
buildings = [[1,4,1],
             [2,7,3],
             [5,9,4],
             [6,10,2],
             [8,15,3],
             [11,13,5],
             [12,18,1],
             [14,17,2],
            ]
```

```python
def plot_buildings(buildings):
    fig1 = plt.figure()
    ax1 = fig1.add_subplot(111)

    ht_max = max([b[2] for b in buildings]) + 1.0
    wd_max = max([b[1] for b in buildings]) + 1.0

    for left, right, ht in buildings:
        ax1.add_patch(
            patches.Rectangle(
                (left/wd_max,0.0),   # (x,y)
                (right-left)/wd_max,          # width
                ht/ht_max,          # height
                fill=False
            )
        )
plot_buildings(buildings)
```

```python
lefts = sorted(buildings, key=lambda b: b[0])
rights = sorted(buildings, key=lambda b: b[1])
n = len(buildings)
h = MaxHeap(2*n)
while len(lefts)>0 or len(rights)>0:
    l = lefts[0] if len(lefts)>0 else [100000, None, None]
    r = rights[0] if len(rights)>0 else [None, None, None]
    print('l',l[0],' r', r[1])
    if l[0] <= r[1]:
        h.insert(l[2])
        lefts.remove(l)
        print('insert', l, 'max:', h.array[0])
    else:
        h.remove(r[2])
        rights.remove(r)
        print('remove', r, 'max:', h.array[0])
```
