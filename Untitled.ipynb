```python
def plot_errors(err_stats, title, units, skip_max=False, skip_sd=False):
    xval = 0
    f, (ax1, ax2) = plt.subplots(2, 1)
    axs = [ax1, ax2]
    mdict = {'min': '*g', 'max': '*r', '99%': 'og', '95%': 'or', '90%':'oy', 'sd': '.g', 'mean': '.r', 'median': '.c'}
    for ix_plot, (err_metric, errors) in enumerate(err_stats.items()):
        if err_metric=='rel-abs-err':
            continue
        for err_type, err_val in errors.items():
            #if skip_if(err_metric, err_type, skip_max, skip_sd):
            #    continue
            axs[ix_plot].plot(err_val, 0 ,mdict[err_type], label=err_type, markersize=10)
            #axs[ix_plot].text(err_val, 0.003 ,err_type + ": " + ("%.2f" % err_val))
        if err_metric=='abs-err':
            axs[ix_plot].set_title(err_metric + " (in "+units+")")
        else:
            axs[ix_plot].set_title(err_metric)
        if err_metric=='abs-err':
            plt.ylabel("error in " + units)
        xlims = axs[ix_plot].get_xlim()
        axs[ix_plot].set_xlim([xlims[0]-5, xlims[1]+5])
    handles = [mlines.Line2D([], [], marker=v[0],color=v[1], markersize=15, label=k, linewidth=0,
                            ) \
               for (k,v) in mdict.items()]
    plt.legend(loc=(1,0.5), handles=handles)
    plt.suptitle(title)
    plt.show()
plot_errors(err_stats, 'yo', 'hrs', skip_max=True, skip_sd=True)
```
