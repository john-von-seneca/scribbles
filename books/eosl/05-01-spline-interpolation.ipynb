source: [Spline Interpolation](https://www.wikiwand.com/en/Spline_interpolation#/Algorithm_to_find_the_interpolating_cubic_spline)

A third order polynomial q(x) for which

$$
\arrthree{
q(x_1)  &= y_1 \\
q(x_2)  &= y_2 \\
q'(x_1) &= k_1 \\
q'(x_2) &= k_2 \\
}
$$

A polynomial satisfying the above constraints is
$$
q =(1-t)y_1 + ty_2+ t(1-t) (a(1-t) + bt)
$$
where
$$
\arrthree{
t &= \ds \frac{x-x_1}{x_2-x_1} \\
a &=  k_1 (x_2 - x_1)-(y_2 - y_1) \\
b &= -k_2 (x_2 - x_1)+(y_2 - y_1) \\
}
$$

As
$$
\arrthree{
q' &= \ds \frac{d q}{d x} = \frac{d q}{d t} \frac{d t}{d x} = \frac{d q}{d t} \frac{1}{x_2-x_1} \\
q'' &= 
\ds \frac{d^2 q}{d t^2} \frac{1}{(x_2-x_1)^2}
}
$$

$$
\arrthree{
\ds \frac{dq}{dt} &= 
(y_2-y_1) +(1-2t) (a(1-t) + b t) + t(1-t) (b-a) \\
\ds \frac{d^2q}{dt^2}
&=
-2(a(1-t) + bt) + (1-2t)(b-a) + (b-a)(1-2t)
\\ &=
2 (b-2a+(a-b)3t)
}
$$

Hence, the second derivatives at $x_1$ and $x_2$ are 

$$
\arrthree{
q''(x_1) &= \ds 2 \frac{b-2a}{(x_2-x_1)^2}
\\
q''(x_2) &= \ds 2 \frac{a-2b}{(x_2-x_1)^2}
}
$$

If now $(x_i, y_i), i = 0, 1, \cdots, n$ are n + 1 points and

$$
q_i = (1-t) y_{i-1} + ty_i + t(1-t)(a_i (1-t) + b_i t)
$$
for $i = 1,\cdots,n$ and $\ds t = \frac{x-x_{i-1}}{x_i - x_{i-1}}$

$$
\arrthree{
a_i &= k_{i-1}(x_i-x_{i-1})-(y_i - y_{i-1}) \\
b_i &= -k_i(x_i-x_{i-1})+(y_i - y_{i-1}) \\
}
$$

for i = 1, ..., n where
$$
\arrthree{
k_0 &= q_1'(x_0) \\
k_i &= q_i'(x_i)=q_{i+1}'(x_i) \qquad i=1,\dotsc ,n-1 \\
k_n &= q_n'(x_n) \\
}
$$

if we want the second derivatives to be equal, we have
$$
q''_{i}(x_i) = q''_{i+1}(x_i)
$$

Lets consider i=0,1,2
$$
\arrthree{
\ds\frac{a_1 - 2 b_1}{(x_1 - x_0)^2}
&=
\ds\frac{b_2 - 2 a_2}{(x_2 - x_1)^2}
}
$$

$$
\arrthree{
a_1 - 2b_1 &=
k_0 (x_1-x_0) + 2k_1(x_1-x_0) - 3(y_1-y_0)\\
b_2 - 2a_2 &=
-k_2(x_2-x_1) - 2k_1(x_2-x_1) + 3(y_2-y_1)
}
$$

sub into the equality and move x-terms to LHS and y-terms to RHS and then, on the LHS, separating the k terms, we have
$$
\frac{k_0}{x_1-x_0} 
+ 2k_1 \Paran{\recp{x_1-x_0}+\recp{x_2-x_1}}
+ \frac{k_2}{x_2-x_1}
\\ =
3 \Paran{
\frac{y_2 - y_1}{(x_2-x_1)^2}
+ \frac{y_1 - y_0}{(x_1-x_0)^2}
}
$$

That is
$$
\frac{k_{i-1}}{x_i-x_{i-1}} +
\Paran{
  \frac{1}{x_i-x_{i-1}}+
  \frac{1}{x_{i+1}-x_i}
}
2k_i
+
\frac{k_{i+1}}{x_{i+1}-x_i}
\\=
3
\Paran{
  \frac{y_i - y_{i-1}}{(x_i-x_{i-1})^2}
  +\frac{y_{i+1} - y_i}{(x_{i+1}-x_i)^2}
}
$$

We have n-1 equations of this sort.

Then we make $\ds q''_1(x_0) = q''_n(x_n) = 0$



$$
\arrthree{
q''_1(x_0) &= 0 \\
\ds \recp{(x_1-x_0)^2} \Paran{-k_1(x_1-x_0) - 2k_0(x_1-x_0) + 3(y_1-y_0)}
&= 0 \\
\ds \frac{2}{x_1-x_0} k_0 + \recp{x_1-x_0}k_1 &=
\ds 3\frac{y_1-y_0}{(x_1-x_0)^2}
}
$$

$$
\arrthree{
q''_n(x_n) &= 0 \\
\ds \recp{(x_n-x_{n-1})^2}
~\Brace{
  k_{n-1}(x_n-x_{n-1}) + 2k_n(x_n-x_{n-1}) - 3(y_n-y_{n-1})
}
&= 0 \\
\ds \frac{1}{x_n-x_{n-1}} ~k_{n-1} +
\frac{2}{x_n-x_{n-1}} ~k_n&=
\ds 3 \frac{y_n-y_{n-1}}{(x_n-x_{n-1})^2}.
}
$$

This is equivalent to a tridiagonal linear system (fav of Gilbert Strang)

$$
\mat{
a_{11} & a_{12} & 0      & 0       & 0      & \cdots \\
a_{21} & a_{22} & a_{23} & 0       & 0      & \cdots \\
0      & a_{32} & a_{33} & a_{34}  & 0      & \cdots \\
\vdots & \ddots & \vdots & \vdots  & \vdots & \vdots \\
0      & \cdots & \cdots & 0  & a_{(n-1)n} & a_{nn}  \\
}
\mat{k_0 \\ \vdots \\ k_n}
=
\mat{
b_1 \\
b_2 \\
b_3 \\
}
$$

That is
$$
\arrthree{
\Ab \kb &= \bb \\
\kb &= \inv{\Ab} \bb
}
$$

Example for n=3

$$
\mat{
\ds\frac{2}{x_1-x_0} & \ds\recp{x_1-x_0} & 0
\\
\ds\recp{x_1-x_0} &
\ds 2 \Paran{ \recp{x_1-x_0}+\recp{x_2-x_1} } &
\ds\recp{x_2-x_1}
\\
0 & \ds\recp{x_2-x_1} & \ds\frac{2}{x_2-x_1}
\\
}
\mat{k_0 \\ k_1 \\ k_2}
=
\mat{
\ds 3 \frac{y_1-y_0}{(x_1-x_0)^2}
\\
\ds 3 \Paran{
  \frac{y_1 - y_0}{(x_1-x_0)^2}+
  \frac{y_2 - y_1}{(x_2-x_1)^2}
}
\\
\ds 3 \frac{y_2-y_1}{(x_2-x_1)^2}
}
$$
